```python
 df_population = pd.read_csv('..\\csv_data\\population_total.csv', encoding='latin-1')
```

```python
df_population
```

```python
# drop null values
df_population.dropna(inplace=True)
```

```python
df_pivot_pop = df_population.pivot(index='year', columns='country', values='population')
```

```python
df_pivot_pop = df_pivot_pop[['United States', 'India', 'China', 'Indonesia', 'Brazil']]
df_pivot_pop
```

# Lineplot

```python
df_pivot_pop.plot(kind='line', xlabel='Year', ylabel='Population',
                 title=' Population (1955-2020)', figsize=(8,4))
```

# Barplot

```python
df_pivot_pop2020 = df_pivot_pop[df_pivot_pop.index.isin([2020])]
df_pivot_pop2020
```

```python
# transpose 
df_pivot_pop2020 = df_pivot_pop2020.T
df_pivot_pop2020
```

```python
df_pivot_pop2020.plot(kind='bar', color='orange')
```

# Barplot grouped by 'n' variables
